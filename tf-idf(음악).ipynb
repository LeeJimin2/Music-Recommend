```python
data = pd.read_csv("멜론TOP100 가사.csv")
```

```python
data
```

```python
data["제목"] = data["제목"].apply(str)
```

```python
data["가사"] = data["가사"].apply(str)
```

```python
X = np.array(data.가사)
```

```python
test_data = X
model = SentenceTransformer('distilbert-base-nli-mean-tokens')
embeddings = model.encode(test_data, show_progress_bar=True)
```

```python
embed_data = embeddings
```

```python
X = np.array(embed_data)
n_comp = 5
pca = PCA(n_components = n_comp)
pca.fit(X)
pca_data = pd.DataFrame(pca.transform(X))
pca_data.head()
```

```python
#all
cos_sim_data = pd.DataFrame(cosine_similarity(X))
def give_recommendations(index,print_recommendation = False,print_recommendation_plots= False,print_genres =False):
    index_recomm =cos_sim_data.loc[index].sort_values(ascending=False).index.tolist()[1:6]
    music_recomm =  data['제목'].loc[index_recomm].values
    result = {'music':music_recomm,'Index':index_recomm}
    if print_recommendation==True:
        print('The listened music is this one: %s \n'%(data['제목'].loc[index]))
        k=1
        for music in music_recomm:
            print('The number %i recommended musics is this one: %s \n'%(k,music))
            k=k+1
    if print_recommendation_plots==True:
        print('The plot of the listened music is this one:\n %s \n'%(data['가사'].loc[index]))
        k=1
        for q in range(len(music_recomm)):
            plot_q = data['가사'].loc[index_recomm[q]]
            print('The plot of the number %i recommended musics is this one:\n %s \n'%(k,plot_q))
            k=k+1
    return result
```

```python
give_recommendations(0, True)
```

```python
give_recommendations(0, False, True)
```

```python
recomm_list = []
for i in range(len(X)):
    recomm_i = give_recommendations(i)
    recomm_list.append(recomm_i['music'])
recomm_data = pd.DataFrame(recomm_list,columns=['First Recommendation','Second Recommendation','Third Recommendation','Fourth Recommendation','Fifth Recommendation'])
recomm_data['listened music'] = data['제목']
recomm_data = recomm_data[['listened music','First Recommendation','Second Recommendation','Third Recommendation','Fourth Recommendation','Fifth Recommendation']]
```
